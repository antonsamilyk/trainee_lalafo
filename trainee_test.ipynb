```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('table_with_numbers.csv', sep = ',', header = None)
df.head()
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
mask_heatmap = np.triu(df.corr())
sns.heatmap(df.corr(), square=True, mask = mask_heatmap)
```

Звернемо увагу на теплову карту, яка відображає попарну кореляцію наших колонок даних. Бачимо, що група колонок по середині має досить високу кореляцію між собою, що може являти собою відповідний кластер. Також бачимо невеличкі осередки подібних ознак, що також може пояснюватися як осередки певних кластерів.

```python
for col in df.columns:
    plt.figure(figsize = (3,3))
    plt.boxplot(df[col])
    plt.title(col)
```

За допомогою відображення ящика з вусами, можемо наглядно перевірити подібність колонок(ознак). 

```python
for col in df.iloc[:,:20].columns:
    plt.figure(figsize = (3,3))
    plt.hist(df[col])
    plt.title(col)
```

Гістограми перших 20 колонок показують, що дані мають в основному значення які сильно відрізніються між собою, тобто мають бімодальний розподіл.
